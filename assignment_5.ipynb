```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
from sklearn.preprocessing import MinMaxScaler
m_m_s = MinMaxScaler()
from sklearn.preprocessing import MaxAbsScaler
m_a_s = MaxAbsScaler()
from sklearn.preprocessing import RobustScaler
r_s = RobustScaler()
```

```python
url = 'https://raw.githubusercontent.com/MeSamiulIslam/aiQuest/main/supershops.csv'
df = pd.read_csv(url) #df = pd.read_csv(url)
df.head(5) ##to get the last 5 rows
     
```

```python
x = df.drop('Profit', axis=1) #axis=1 for column
```

```python
x.head()
```

#Standard Scaler

```python
x.info()
```

```python
for column in x.columns:
    if not x.dtypes[column] == object:
        x[column] = std.fit_transform(x[[column]])

x.head()
```

#MinMax Scaler

```python
clms = ['Marketing Spend', 'Administration', 'Transport']
```

```python
df[clms] = m_m_s.fit_transform(df[clms])
df.head()
```

#Maximum absolute Scaler

```python
df[clms] = m_a_s.fit_transform(df[clms])
df.head()
```

#Robust Scaler

```python
df[clms] = r_s.fit_transform(df[clms])
df.head()
```
