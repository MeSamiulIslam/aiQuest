#Extra Trees Classifier

```python
from sklearn.ensemble import ExtraTreesClassifier #Decision Tree
```

```python
extra = ExtraTreesClassifier()
```

```python
extra.fit(x,y)
```

```python
extra.feature_importances_ #IG from decision treee
```

```python
feature_importances = extra.feature_importances_ 
feature_importances
```

```python
imp = pd.DataFrame(feature_importances, columns= ['Gain_score'])
```

```python
imp
```

```python
x.columns
```

```python
cols = pd.DataFrame(x.columns, columns = ['Feature_Names'])
cols
```

```python
gains = pd.concat([cols, imp], axis=1)
```

```python
gains
```

```python
gains.nlargest(15,'Gain_score')
```

```python
gains.nsmallest(8, 'Gain_score')#smallest()
```

```python
new_x = gains.nlargest(15, 'Gain_score')
```
